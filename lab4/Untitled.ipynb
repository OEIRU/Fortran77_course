```python
import matplotlib.pyplot as plt
import numpy as np

# Функция для чтения данных из файла
def read_data_from_file(filename):
    N = []
    errors = []
    with open(filename, 'r') as file:
        lines = file.readlines()
        for i in range(0, len(lines), 2):  # Чтение пар строк (N и Error)
            n_line = lines[i].strip()
            error_line = lines[i+1].strip()
            
            # Извлечение значений
            n_value = int(n_line.split()[-1])
            error_value = float(error_line.split()[-1])
            
            N.append(n_value)
            errors.append(error_value)
    return np.array(N), np.array(errors)

# Загрузка данных из файла
filename = 'input.txt'  # Укажите путь к вашему файлу
N, errors = read_data_from_file(filename)

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(N, errors, marker='o', linestyle='-', color='b')

# Логарифмический масштаб для оси Y
plt.yscale('log')

# Подписи осей и заголовок
plt.xlabel('N')
plt.ylabel('Error (log scale)')
plt.title('Error vs N (log scale)')

# Сетка
plt.grid(True, which="both", ls="--")

# Показать график
plt.show()
```
